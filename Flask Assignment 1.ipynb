Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python. It is designed to be simple and lightweight, providing the basic tools and features needed to build web applications. Here are the advantages of using Flask:

Lightweight and easy to learn: Flask has a small and intuitive core, making it easy for developers to understand and get started with. It follows a minimalistic approach, allowing developers to add only the features they need.

Flexible and customizable: Flask gives developers the flexibility to choose the components they want to use, allowing for greater customization. It doesn't impose any specific project structure or database, giving developers the freedom to design their applications as they see fit.

Integrated development server: Flask comes with a built-in development server, which makes it convenient to test and debug applications without the need for additional setup.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask, render_template, request, jsonify

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

@app.route('/')
def company_details():
    return '''<pre>
              Company Name: ABC Corporation \n
              Location: India \n
              Contact Detail: 999-999-9999</pre>'''
if __name__=="__main__":
     app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for the specified endpoint by taking into account the routing rules defined in the Flask application. The url_for() function takes the endpoint name as an argument and can also accept variable arguments to build dynamic URLs.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    # Generate URL for the 'hello' endpoint
    hello_url = url_for('hello')
    return f"Go to <a href='{hello_url}'>Hello</a>"

@app.route('/hello')
def hello():
    return "Hello, world!"

if __name__=="__main__":
     app.run(host="0.0.0.0")
```
