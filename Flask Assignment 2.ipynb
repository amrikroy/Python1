Q1. Explain GET and POST methods.

GET: The GET method is used to retrieve or fetch data from a server. When a GET request is made, the client (such as a web browser) sends a request to the server to retrieve a specific resource (e.g., a webpage or an image). The data is appended to the URL as query parameters, visible in the address bar of the browser. GET requests are generally considered to be idempotent, meaning that making the same request multiple times will have the same result.
Example: When you enter a URL in your browser or click on a hyperlink, a GET request is made to retrieve the corresponding webpage

The POST method is used to submit data to be processed by the server. When a POST request is made, the client sends data to the server in the body of the request, rather than appending it to the URL. This method is often used for creating or updating resources on the server. Unlike GET requests, POST requests are not idempotent, meaning that making the same request multiple times may result in different outcomes.
Example: When you submit a form on a website, such as a registration form, the data you enter is sent to the server using a POST request.

Q2. Why is request used in Flask?

In Flask, the request object is used to handle incoming HTTP requests made to a Flask application. It provides access to various attributes and methods that allow developers to retrieve and manipulate data sent by the client.

Retrieving data: The request object allows you to access data sent by the client in the form of query parameters, form data, JSON payloads, or file uploads. It provides methods to extract this data, making it easy to process and utilize in your Flask application.

Handling HTTP methods: The request object helps determine the HTTP method used in the request (GET, POST, PUT, DELETE, etc.). It allows you to perform different actions based on the method, enabling you to create dynamic and interactive applications.

URL handling: The request object provides methods to work with URLs. It allows you to retrieve the URL of the current request, extract URL parameters, or construct URLs for redirection or other purposes.

Q3. Why is redirect() used in Flask?

In Flask, the redirect() function is used to redirect the client's web browser to a different URL. It is commonly used when you want to redirect users to another page after performing a certain action or when handling certain conditions in your Flask application.

Route handling: In some cases, you may want to redirect users to different routes based on certain conditions. For example, if a user attempts to access a restricted page without authentication, you can use redirect() to send them to the login page. It allows you to dynamically control the flow of your application and direct users to the appropriate pages.

URL manipulation: redirect() can also be used to manipulate URLs. You can modify query parameters, add or remove parts of the URL, or construct entirely new URLs to redirect the user to. This flexibility allows you to create dynamic and personalized redirections based on the specific requirements of your application.

Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are files that contain the HTML structure and dynamic content placeholders of a web page. They allow developers to separate the presentation logic from the application logic, making it easier to manage and update the user interface.

Templates in Flask: Templates are typically stored in a dedicated directory within the Flask application, often named "templates." They are written in HTML, but they can also contain placeholders for dynamic content using templating languages such as Jinja2, which is the default templating engine in Flask. Templates allow you to define the structure, layout, and design of web pages while keeping the content separate from the code.

render_template() function: The render_template() function is used in Flask to render a specified template file and return it as an HTTP response.
